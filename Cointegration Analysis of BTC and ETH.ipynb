```python
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller

dropcol_names = ['Open', 'High', 'Low', 'Adj Close', 'Volume']
btc_price = btc.drop(dropcol_names, axis=1)
eth_price = eth.drop(dropcol_names, axis=1)

btc_pct2020 = btc_price.pct_change()
eth_pct2020 = eth_price.pct_change()

btc_pct2020.dropna(inplace=True)
eth_pct2020.dropna(inplace=True)

eth_price = sm.add_constant(eth_price)

result = sm.OLS(btc_price,eth_price).fit()

#compute ADF 
```

```python
# Does the BTC / ETH prices follow a random walk?
btcresults = adfuller(btc['Close'])
ethresults = adfuller(eth['Close'])

print('The p-value of the test on BTC prices is ' + str(btcresults[1]))
print('The p-value of the test on ETH prices is ' + str(ethresults[1]))
```

```python
b = result.params[1]
adf_stats = adfuller(btc_price['Close'] - b * eth_price['Close'])
print('The pvalue for the ADF test is ', adf_stats[1])
```

```python

#Let's look at the 2016 prices for BTC and ETH
```

```python
btc_file = 'BTC-USD-2016.csv'

btc_path = data_path + btc_file 

btc2016 = pd.read_csv(btc_path, index_col='Date', parse_dates=date_cols)

eth_file = 'ETH-USD-2016.csv'

eth_path = data_path + eth_file 

eth2016 = pd.read_csv(eth_path, index_col='Date', parse_dates=date_cols)
```

```python
btc_price2016 = btc2016.drop(dropcol_names, axis=1)
eth_price2016 = eth2016.drop(dropcol_names, axis=1)

btc_pct2016 = btc_price2016.pct_change()
eth_pct2016 = eth_price2016.pct_change()

eth_pct2016.dropna(inplace=True)
btc_pct2016.dropna(inplace=True)


eth_price2016 = sm.add_constant(eth_price2016)

result = sm.OLS(btc_price2016,eth_price2016).fit()
```

```python
b = result.params[1]
adf_stats = adfuller(btc_price2016['Close'] - b * eth_price2016['Close'])
print('The pvalue for the ADF test is ', adf_stats[1])
```

```python
eth_pct2016 = sm.add_constant(eth_pct2016)
eth_pct2020 = sm.add_constant(eth_pct2020)

result2016 = sm.OLS(btc_pct2016, eth_pct2016).fit()
result2020 = sm.OLS(btc_pct2020, eth_pct2020).fit()



b16=result2016.params[1]
b20 = result2020.params[1]

adf_stats_2016 = adfuller(btc_pct2016['Close'] - b16*eth_pct2016['Close'])
adf_stats_2020 = adfuller(btc_pct2020['Close'] - b20*eth_pct2020['Close'])

print("the p value for the 2016 ADF test is ",adf_stats_2016[1])
print("the p value for the 2020 ADF test is ",adf_stats_2020[1])

    
```
